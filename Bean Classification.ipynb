#### TREATING OUTLIER USING IQR

```python
# first spli into X and y
X=dt.drop("Class",axis=1)
y=dt["Class"]
```

```python
# Train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42,stratify=y)
```

```python
# Calculate IQR on train only 

Q1=X_train.quantile(0.25)
Q3=X_train.quantile(0.75)

IQR=Q3-Q1

lower_bound=Q1-3*IQR
upper_bound=Q3+3*IQR

# filtering outlier for TRAIN

mask_train=~((X_train<lower_bound)|(X_train>upper_bound)).any(axis=1)

X_train_outlier=X_train[mask_train]
y_train_outlier=y_train[mask_train]

# filtering outlier for TEST

mask_test=~((X_test<lower_bound)|(X_test>upper_bound)).any(axis=1)

X_test_outlier=X_test[mask_test]
y_test_outlier=y_test[mask_test]
```

```python
print("Original Train:", X_train.shape)
print("Train after outlier removal:", X_train_outlier.shape)

print("Original Test:", X_test.shape)
print("Test after outlier removal:", X_test_outlier.shape)
```

#### Check for skewness  and treat it 

```python
X_train_skew=X_train_outlier.copy()
X_test_skew=X_test_outlier.copy()

# check for skew 
skewness=X_train_skew.skew()
print(skewness)
```

```python
# for positive skewness features 
skewed_features_positive=skewness[abs(skewness) > 0.5].index
X_train_skew[skewed_features_positive]=np.log1p(X_train_skew[skewed_features_positive])
X_test_skew[skewed_features_positive]=np.log1p(X_test_skew[skewed_features_positive])



```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score

results = []

def add_result(name, y_train, y_train_pred, y_test, y_test_pred):
    train_acc = accuracy_score(y_train, y_train_pred)
    test_acc = accuracy_score(y_test, y_test_pred)
    f1 = f1_score(y_test, y_test_pred, average='weighted')
    overfit = "Yes" if train_acc - test_acc > 0.05 else "No"
    
    results.append([name, round(train_acc,4), round(test_acc,4), round(f1,4), overfit])


# ===== ADD YOUR MODELS HERE =====

add_result("Logistic Regression", y_train_smote, y_pred_train_lr, y_test_encoded, y_pred_test_lr)

add_result("Decision Tree", y_train_smote, y_pred_train_dt, y_test_encoded, y_pred_test_dt)

add_result("Random Forest", y_train_smote, y_pred_train_rf, y_test_encoded, y_pred_test_rf)

add_result("SVM", y_train_smote, y_pred_train_svm, y_test_encoded, y_pred_test_svm)

add_result("KNN", y_train_smote, y_pred_train_knn, y_test_encoded, y_pred_test_knn)

add_result("Gradient Boosting", y_train_smote, y_pred_train_gbc, y_test_encoded, y_pred_test_gbc)

add_result("CatBoost", y_train_smote, y_pred_train_cat, y_test_encoded, y_pred_test_cat)

add_result("CatBoost_tuned", y_train_smote,y_pred_train_cat2, y_test_encoded, y_pred_test_cat2)


# ===== CREATE TABLE =====

df = pd.DataFrame(results, columns=["Model","Train Acc","Test Acc","F1 Score","Overfitting"])
df = df.sort_values("F1 Score", ascending=False).reset_index(drop=True)

print(df)

# Best Model
print("\n🔥 Best Model:", df.iloc[1]["Model"])
```
